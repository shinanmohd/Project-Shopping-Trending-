```python
plt.figure(figsize=(8,6))
plt.bar(trend['Item Purchased'], trend['Purchase Amount (USD)'], color='skyblue')
plt.xlabel('Item Purchased')
plt.ylabel('Purchase Amount (USD)')
plt.title('Purchase Amount by Item Purchased')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show();
```

```python
payment_method_counts = trend['Payment Method'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(payment_method_counts, labels=payment_method_counts.index,explode=(0.1,0,0,0,0,0),autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral', 'gold'])
plt.title('Payment Method Distribution', fontsize=14)
plt.show()
```
